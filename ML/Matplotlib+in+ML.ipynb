## Matplotlib


Matplotlib is a plotting library like GNUplot. The main advantage towards GNUplot is the fact that Matplotlib is a Python module.

```python
import matplotlib.pyplot as plt
plt.plot([-1,-4.5,16,23])
```

```python
plt.show()
```

```python
plt.plot([-1, -4.5, 16, 23], "ob")
```

```python
plt.show()
```

```python
# our X values:
days = list(range(0, 22, 3))
print(days)
# our Y values:
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
plt.plot(days, celsius_values)
```

```python
plt.show()
```

```python
plt.plot(days, celsius_values, 'ob')
```

```python
plt.show()
```

## Labels on Axes

```python
days = list(range(1,9))
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
plt.plot(days, celsius_values)
plt.xlabel('Day')
plt.ylabel('Degrees Celsius')
```

```python
plt.show()
```

```python
days = list(range(1,9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]
plt.xlabel('Day')
plt.ylabel('Degrees Celsius')
plt.plot(days, celsius_min,
         days, celsius_min, "oy",
         days, celsius_max, 
         days, celsius_max, "or")
```

```python
plt.show()
```
